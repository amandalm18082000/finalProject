## **Preprocessing**

**Convert object to numeric datatypes**

```python
columns = ['AQI index','Dominent pollutant', 'CO', 'NO2', 'O3', 'Pressure',\
       'PM10', 'PM2.5', 'SO2']
```

```python
data[columns] = data[columns].astype(str).apply(lambda x: pd.to_numeric(x.str.replace('\D','',regex=True)))
```
